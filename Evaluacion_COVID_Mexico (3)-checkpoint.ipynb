## Análisis numérico de los datos
### 1-Número de positivos y negativos
&emsp;&emsp;Una vez realizado el análisis previo de los datos, se procede a comprobar el número de positivos detectados, el número de negativos detectados y el porcentaje de positivos sobre el total de tests, ya que, un porcentaje alto indica que se están realizando menos tests de los necesarios. Para contabilizar los registros se utiliza la columna **ID_REGISTRO** por considerarse que es la clave en el dataframe principal.

&emsp;&emsp;Tal como se puede observar en los resultados, parece que se están realizando un número suficiente de tests de COVID entre la población mexicana y, a pesar de obtener un número relativamente alto de positivos (24,905), solamente representan el **24.9%** de los tests realizados.

```python
#Obtención del número de resultados POSITIVOS y NEGATIVOS
positivos = df[df["RESULTADO"] == 1]["ID_REGISTRO"].count()
totalTests = df["ID_REGISTRO"].count()
negativos = totalTests - positivos
print("Número de positivos:\t\t", f'{positivos:,d}', "\nNúmero de negativos:\t\t",f'{negativos:,d}',\
      "\nTotal de tests realizados:\t",f'{totalTests:,d}', \
     "\nPorcentaje de positivos:\t",f'{(positivos/totalTests)*100:,.2f}%')
```

### 2-Pacientes que han necesitado ingreso en la UCI

&emsp;&emsp;El siguiente punto a evaluar es el número de personas que han necesitado ingresar en la UCI así como qué porcentaje representa sobre el total de positivos que se han dado.

&emsp;&emsp;Tal como se puede comprobar, solamente han necesitado ingresar en la UCI **2,250** personas, resultando poco más del **9%** de los infectados.

&emsp;&emsp;De este dato se deduce que el nivel de gravedad del COVID19 en México es relativamente bajo.

```python
#Obtención del número de personas que han necesitado ser ingresadas en la UCI
uci = df[(df["RESULTADO"] == 1) &(df["UCI"] == 1)]["ID_REGISTRO"].count()
print(f'{uci:,d} personas infectadas de COVID19 han necesitado ingresar en la UCI, representando el '\
      f'{(uci/positivos)*100:,.2f}% de los infectados')
```

### 3-Número de pacientes que han fallecido:

&emsp;&emsp;Una vez analizado el número de personas infectadas por COVID19 que han necesitado ingresar en la UCI, se procede a analizar el número de personas fallecidas por COVID19. Para ello, se estudia la columna **FECHA_DEF** que nos indica la fecha de defunción del paciente. Si el valor de este campo es **9999-99-99** significa que, a fecha de extracción de los datos, el paciente estaba vivo. Además, nos interesa solamente aquellas defunciones provocadas por el virus.

&emsp;&emsp;Tal como se puede ver, 2,270 personas han fallecido por COVID19, representando una tasa de mortalidad del **9.12%**. Se puede comprobar que es una tasa relativamente alta respecto a la tasa de mortalidad media en el mundo (que, a fecha 25 de julio, según las publicaciones oficiales, es del **4.06%**).

&emsp;&emsp; Asimismo, se puede comprobar que ha habido una gran cantidad de fallecimientos de pacientes que no han podido ingresar en una UCI (2,270 fallecidos sobre 2,250 personas ingresadas en la UCI) con lo que se puede extrapolar que la sanidad en México no llega a una parte bastante importante de la población.

```python
fallecimientos = df[(df["FECHA_DEF"] != '9999-99-99') & \
                   (df["RESULTADO"] == 1)]["ID_REGISTRO"].count()
print(f'{fallecimientos:,d} han fallecido por coronavirus, resultando una tasa de mortalidad de: '\
      f'{(fallecimientos/positivos)*100:,.2f}% del total de pacientes infectados por COVID19')
```
